```python
import os 
from urllib.request import urlopen

from bs4 import BeautifulSoup 
```

```python
page = "https://www.business-school.ed.ac.uk/"
website = urlopen(page)
soup = BeautifulSoup(website, 'html.parser')
```

```python
print(soup)
```

```python
headertags = soup.find(id = "contactHeader")
print(headertags)
print(headertags.text)
```

```python
tagsinContact = headertags.findChildren()
for tag in tagsinContact:
    print("code : ", tag ,"Text: ", tag.text)
```

```python
class_elements = soup.find_all("div", {"class": "jumbotron jumbotron-fluid"})
for elements in class_elements:
    print( " Text: ", elements.text)
```

```python
new_div = soup.find_all("div",{"class":"views-rows"})
row_num = 0 
for new in new_div:
    for row in new.find_all("div"):
        print("ROW: "+str(row_num))
        row_num = row_num+1
    
```

```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
```

```python
import sys
def get_a_browser():
    if sys.platform.startswith('win32') or sys.platform.startswith('cygwin'):
        return webdriver.Chrome() # windows
    else:
        return webdriver.Chrome('chromedriver') # mac
```
