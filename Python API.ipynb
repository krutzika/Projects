```python
def remove_stopwords(additionalList,falseList):
        
        stopwordList = []
        stopwordList = stopwords.words('english')
        stopwordList = stopwordList + additionalList
        stopwordsnewList = []
        for x in falseList:
            for y in stopwordList:
                if x != y :
                    stopwordsnewList.append(y)
        return stopwordsnewList
```

```python
def top_n_ngram(corpus,n = None,ngram = 1, words_stop = stopwords.words('english')):
        vec = CountVectorizer(stop_words = words_stop,ngram_range=(ngram,ngram)).fit(corpus)
        bag_of_words = vec.transform(corpus) #Have the count of  all the words for each review
        sum_words = bag_of_words.sum(axis =0) #Calculates the count of all the word in the whole review
        words_freq = [(word,sum_words[0,idx]) for word,idx in vec.vocabulary_.items()]
        words_freq = sorted(words_freq,key = lambda x:x[1],reverse = True)
        return words_freq[:n]
```

```python
def plot_histogram_words(data, stopword):
        common_words = top_n_ngram(data['description_string'], 20, 3, stopword)
        df = pd.DataFrame(common_words, columns = ['ReviewText' , 'count'])
        plt.style.use('default')
        plt.figure(figsize =(20,10))
        df.groupby('ReviewText').sum()['count'].sort_values(ascending=True).plot(
        kind='barh', title='Top 20 unigrams in review after removing stop words')
```

```python
def feature_extraction(df, column, column_name):
    df[column] = pd.to_datetime(df[column], errors='coerce')
    df[f'day_{column_name}'] = df[column].dt.day
    df[f'month_{column_name}'] = df[column].dt.month
    df[f'year_{column_name}'] = df[column].dt.year
    return df
    
```

```python
def plot_dateTime(x_value,xlabel, ylabel, data ):
    rush_month_fig = plt.figure(figsize=(15,5))
    sns.countplot(x=x_value, data=data)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.show()
    
```

```python
def filling_link(link_data):
    new_link=link_data['links.filing_history']
    new_link_chargers = new_link.dropna()
    new_link_chargers_list = new_link_chargers.tolist()
    return new_link_chargers_list
```

### Business Question 2: Mini-report and visualisation:

<b>Problem Statement</b>: UK financial services wants to evaluate the effect of food price spike and crisis on food related companies post global financial recession in 2007-2008.

<b>Solution</b>: Despite 2012-2013 having the highest food price spike in the UK since 2007, this is also the time with highest new food services/retailers establishment after 2007. With London and Manchester topping this list. It is noted from Graph 1 that around 50% of these companies closed in the period 2015-2016. With highest closed down happening in London, the filling history of these companies denotes that 60% of these companies were filed with gazette compulsory notice before termination, which indicates that these companies should terminate within 3 months of issuing due to non-payment of taxes, accounting and so on.

Analysing the time during which this notice was filled, it is observed from graph 2 that majority of these notices took place during the second half of the year 2015 and early 2016. 

<b>Conclusion</b>: The rise of food spike made it difficult for many of the companies which started in 2012-2013 to sustain for more than 2 years of time. Chinese financial crisis that took place during mid 2015, might have affected these food companies. Providing tax waivers and relaxation of accounting laws during the time of crisis is important to tackle a recession.
